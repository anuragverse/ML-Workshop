```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

df = pd.read_csv('Income.csv')
features = ['Annual Income(rs)', 'Spending Score(1-100)']
X = df[features]


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)

df['Cluster'] = kmeans.labels_

print(df)

plt.scatter(df['Annual Income(rs)'], df['Spending Score(1-100)'], c=df['Cluster'], cmap='viridis', label='Customers')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='.', label='Cluster Centers')
plt.title('KMeans Clustering')
plt.xlabel('Annual Income(rs)')
plt.ylabel('Spending Score(1-100)')
plt.legend()
plt.show()
```
