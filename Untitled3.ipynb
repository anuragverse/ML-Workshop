```python
!pip install tensorflow==2.12
!pip install deepface==0.0.79

from deepface import DeepFace
import matplotlib.pyplot as plt
import cv2

# Load the image
img_path = '/content/Messi1.jpg'
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB

# Analyze the image using DeepFace
try:
    analysis = DeepFace.analyze(img, actions=['gender'])
    gender = analysis[0]['dominant_gender']  # Get the dominant gender prediction

    # Display the image and gender prediction
    plt.imshow(img)
    plt.title(f"Detected Gender: {gender}")
    plt.axis('off')
    plt.show()
except ValueError:
    print("No face detected in the image.")
```
