```python

# Install necessary library
!pip install opencv-python
# Upload image
from google.colab import files
uploaded = files.upload()
# Import libraries
import cv2
from matplotlib import pyplot as plt
# Read and display the image
image = cv2.imread(list(uploaded.keys())[0])
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.show()
# SIFT feature extraction and visualization
sift = cv2.SIFT_create()
keypoints, descriptors = sift.detectAndCompute(gray, None)
image_sift = cv2.drawKeypoints(image, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(cv2.cvtColor(image_sift, cv2.COLOR_BGR2RGB))
plt.title("SIFT Keypoints")
plt.show()
# HOG feature extraction and visualization
hog=cv2.HOGDescriptor()
h = hog.compute(gray)
plt.plot(h)
plt.title("HOG Features")
plt.show()
```
