```python
!pip install tensorflow==2.12
!pip install deepface==0.0.79

from deepface import DeepFace
import matplotlib.pyplot as plt
import cv2

# Load the image
img_path = '/content/Messi1.jpg'
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB

# Analyze the image using DeepFace for age, gender, and emotion
try:
    analysis = DeepFace.analyze(img, actions=['age', 'gender', 'emotion'])  # Include age and emotion

    # Get the results
    age = analysis[0]['age']
    gender = analysis[0]['dominant_gender']
    emotion = analysis[0]['dominant_emotion']

    # Display the image and predictions
    plt.imshow(img)
    plt.title(f"Age: {age}, Gender: {gender}, Emotion: {emotion}")  # Include all predictions in the title
    plt.axis('off')
    plt.show()
except ValueError:
    print("No face detected in the image.")
```
